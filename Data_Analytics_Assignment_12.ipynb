1. Read the Bike Details dataset into a Pandas DataFrame and display its first 10 rows.
(Show the shape and column names as well.)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/drive/MyDrive/Dataset/BIKE DETAILS.csv")
print("First 10 rows of the dataset:")
display(df.head(10))
print("\nShape of the dataset (rows, columns):")
print(df.shape)
print("\nColumn names in the dataset:")
print(df.columns.tolist())
```

2. Check for missing values in all columns and describe your approach for handling them.

```python
print("Missing values in each column:")
print(df.isnull().sum())

df_clean = df.copy()
for col in df_clean.columns:
    if df_clean[col].dtype == 'object':
        df_clean[col].fillna(df_clean[col].mode()[0], inplace=True)
    else:
        df_clean[col].fillna(df_clean[col].median(), inplace=True)

print("\nMissing values after handling:")
print(df_clean.isnull().sum())
```

Approach for Handling Missing Values:-
The strategy for handling missing values depends on the nature of the data, the extent of missingness, and the goals of the analysis. Common approaches include:
  - Deletion:
      - Listwise Deletion: Removing entire rows where any column has a missing value. This is simple but can lead to significant data loss, especially with many missing values or a small dataset.
      - Column Deletion: Removing entire columns that have a high percentage of missing values. This is suitable if a column contains too little information to be useful.
  - Imputation:
  - Replacing missing values with estimated values.
      - Mean/Median/Mode Imputation:
          - Numerical Data: Replace missing values with the mean (for normally distributed data) or median (for skewed data or data with outliers) of the respective column.
          - Categorical Data: Replace missing values with the mode (most frequent value) of the respective column.
      - Forward/Backward Fill: In time-series data, filling missing values with the previous or next observed value.
      - Advanced Imputation Techniques:
          - Regression Imputation: Using a regression model to predict missing values based on other variables in the dataset.
          - K-Nearest Neighbors (KNN) Imputation: Imputing missing values by considering the values of the k-nearest neighbors in the feature space.
          - Multiple Imputation: Creating multiple complete datasets by imputing missing values multiple times, then combining the results.

3. Plot the distribution of selling prices using a histogram and describe the overall trend.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,7))
sns.histplot(df['selling_price'], bins=30, kde=True)
plt.title("Distribution of Selling Prices")
plt.xlabel("Selling Price")
plt.ylabel("Frequency")
plt.show()

print("\nSummary statistics for selling_price:")
print(df['selling_price'].describe())
```

6. Use the IQR method to detect and remove outliers from the km_driven column. Show before-and-after summary statistics.

```python
print("Before removing outliers (km_driven stats):")
print(df['km_driven'].describe())


Q1 = df['km_driven'].quantile(0.25)
Q3 = df['km_driven'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df_no_outliers = df[(df['km_driven'] >= lower) & (df['km_driven'] <= upper)]


print("\nAfter removing outliers (km_driven stats):")
print(df_no_outliers['km_driven'].describe())

print(f"\nRows before: {len(df)}, Rows after: {len(df_no_outliers)}")
```

Most important factors affecting a bike's selling price:-

  - Physical Condition: A clean, well-maintained bike with no cosmetic damage like scratches or dents is more appealing to buyers and will fetch a higher price.
  - Maintenance History and Documentation: A complete and consistent maintenance record shows the bike has been cared for and can increase buyer confidence and the selling price.
  - Brand Reputation: Bikes from well-known and trusted brands generally hold their value better due to their reliability and high demand in the used market.
  - Model Popularity & Demand: The popularity of a specific bike model significantly impacts its resale value; more desirable models command higher prices.
  - Mileage: Lower mileage generally indicates less wear and tear, making the bike more attractive to buyers and contributing to a higher selling price.
  - Aftermarket Accessories: While popular, functional upgrades can enhance a bike's value, highly personalized or niche modifications may detract from the price as they can deter potential buyers.
  - Economic Conditions: Broader economic factors, such as the overall state of the economy and consumer spending power, can influence the demand for used bikes.
